```
請寫一程式，引導使用者輸入中文、英文、數學、和程式 設計四科的成績，並顯示四科的平均成績、GPA以及是否及格
GPA:
80-100分 -> A
70-79分 -> B
60-69分 ->C
低於60分 -> D
範例輸入：
80
80
90
90
範例輸出：

```

````python
"""
GPA:
80-100分 -> A
70-79分  -> B
60-69分  -> C
60分-    -> D
"""
class_list = ['國文','英文','數學','程式設計']
score_list = []
gpa = ""
pas = "pass"
for i in range(4):
    score_list.append(int(input(class_list[i])))
score_avg = (sum(score_list))/4
if score_avg<60:
    gpa = "D"
    pas = "false"
elif score_avg<70 and score_avg>=60:
    gpa = "C"
elif score_avg<80 and score_avg>=70:
    gpa = "B"
elif score_avg>=80:
    gpa = "A"
elif score_avg>100:
    gpa = "AAAAAAAAAAAAAA巨錘瑞斯"
print(f"平均 {score_avg}\n{gpa}\n{pas}")
````

### HW2
```
輸入一個整數N(1 ≤ N ≤ 2 ∗ 10^100)，請計算

S = 1^1 + 2^2 +3^3 + … + N^N

請將S的最後一位數字進行輸出

輸入：

輸入包含數個N，每一行輸入一個N，當N=0時結束

輸出：

每一個N的輸出結果都只有一個數字，該數字為S的最後一位數字

Sample Input
1

2

3

0

Sample Output

1

5

2
```

````python
num_list = []
n = 1
while(int(n)!=0):
    n = input()
    num_list.append(n)

num_list.pop()
print("Output:")
for i in num_list:
    num_sum = 0
    for j in range(int(i)+1):
        if j == 0:
            continue
        num_sum = num_sum + pow(j,j)
    num_sum = [int(x) for x in str(num_sum)]
    print(num_sum[-1])
````

### HW3
```
在這個問題中，你必須分析特定的排序演算法。該演算法通過交換兩個相鄰的序列元素，來處理由n個不同整數所組成的序列，直到該序列以升序排序。

以輸入序列 9 1 0 5 4來說，經過排序後會得到輸出0 1 4 5 9。

你需要做的是計算排序過程中，需要經過幾次交換才能完成排序。

輸入：

輸入包含數個測資，每一組測資的第一行為一個整數n < 500,000，n 為序列的長度，接下來的n 行輸入，每一行的輸入為一個整數a[i] ( 0 ≤ a[i] ≤ 999,999,999 )，i 為第i個序列元素，

若輸入一個序列的長度為n=0，則不進行處理。

輸出：

每一組輸入的序列，需要輸出一個整數op，op為排序完成最少需要經過的交換次數

Sample Input

5

9

1

0

5

4

3

1

2

3

0

Sample Output

6

0

Hint:

try:

    .......

except EOFError:

    pass
```

````python
count_list=[]
n = 1
try:
    while(int(n)!=0):
        n = int(input())
        num_list = []
        for i in range(n):
            num_list.append(input())
        count = 0
        for i in range(len(num_list)-1):
            for j in range(len(num_list)-1):
                if num_list[j]>num_list[j+1]:
                    num_list[j], num_list[j+1] = num_list[j+1], num_list[j]
                    count += 1
        count_list.append(count)
except EOFError:
    pass

for i in range(len(count_list)-1):
    print(count_list[i])
````
